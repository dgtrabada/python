En el criptoanálisis, la técnica de análisis de frecuencias consiste en el aprovechamiento de estudios sobre la frecuencia de las letras o grupos de letras en los idiomas para poder establecer hipótesis para aprovecharlas para poder descifrar un texto cifrado sin tener la clave de descifrado (romper). Es un método típico para romper cifrados clásicos. 
Haz una anailisis de frecuencias para los 10 primeros càpitulos de el libro "el_quijote.txt"

```python
abc="ABCDEFGHIJKLMNÑOPQRSTUVWXYZ"
f=[]
for i in abc:
    f.append(1)
N=0
archivo="el_quijote.txt"
text=open(archivo).readlines()
nmaxlines=len(text)
for line in text:
    for word in line.split():
        for i in word.upper():
            for j in abc:
                a=i.upper().replace("Á","A").replace("É","E").replace("Í","I").replace("Ó","O").replace("Ú","U")
                if a==j :                 
                    f[(abc.index(i))]= f[(abc.index(i))]+1 
                    N=N+1
        
for i in abc:
    f[(abc.index(i))]=f[(abc.index(i))]/N
    
for i in abc:
   print("%.2c  " % i,"%.3f" % f[(abc.index(i))])
```

<b><font color='red'>Generar el ejercicio "El escarabajo de oro"</font></b>

En este cuento de Edgar Allan Poe, el protagonista (el señor Legrand) encuentra un tesoro gracias a que es capaz de descifrar un texto en clave con las pistas suficientes para ir en su busca.  

El señor Legrand consigue descifrar el texto cifrado porque va sustituyendo cada signo (cifra o símbolo) por una letra del abecedario, teniendo en cuenta que los signos que más se repiten deben corresponder, más o menos, a las letras que más se usan en español: la E, la A, la O... Después del mensaje en clave puedes ver, en rojo, el abecedario ordenado por orden de frecuencia: la letra más usada es la E, seguida de la A, etc.

```python
abc="ABCDEFGHIJKLMNÑOPQRSTUVWXYZ"
rem="ESCARBJODFGHIKLMNÑPQTUVWXYZ"
f=[]
for i in abc:
    f.append(1)

texto=""
archivo="El_escarabajo_de_oro_SOL.txt"
text=open(archivo).readlines()
nmaxlines=len(text)
for line in text:
    for word in line.split():
        for i in word.upper():
            sust=False
            for j in abc:
                a=i.upper().replace("Á","A").replace("É","E").replace("Í","I").replace("Ó","O").replace("Ú","U")
                if a==j :                 
                    texto=texto+rem[abc.index(j)]
                    sust=True
            if not sust:
                texto=texto+i 
        texto=texto+' '
    texto=texto+'\n'
        
print(texto)
```
