Representa la función $\frac{D}{(x^2+1)}$ , donde D es el día de tu cumpleaños  
Hazque el título sea tenga una fuente: serif de color blue y tamaño 16

<img src="f1.png" ALIGN="left">


```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(-4, 4, 0.1) 
f = (17/((x**2)+1))

plt.plot(x,f,label=r'$\frac{17}{(x^2+1)}$')
font = {'family':'serif','color':'blue','size':16}
plt.title("Tu nombre y apellido", loc = 'center', fontdict = font)
plt.xlabel("x")
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
```

Representa también la función con D/2, busca y representa también como poner las barras verticales y horizontales

<img src="f2.png" ALIGN="left">

```python
# representa la función D/(x**2+1) , donde D es el día de tu cumpleaños

def f(x,D):
    return D/(x**2+1)

x = np.arange(-4, 4, 0.1)
y2 = []
y=[]
for i in x:
    y.append(f(i,25))
    
for i in x:
    y2.append(f(i,12.5))    
    
plt.plot(x,y, linestyle = '-',linewidth = '1.5',color='g',label='$\\frac{25}{(x^2+1)}$')
plt.plot(x,y2, linestyle = '--',linewidth = '1.5',color='r',label='$\\frac{12.5}{(x^2+1)}$')
#plt.plot(x,x*8,marker = '+', linestyle = '--',linewidth = '1.5',color='r',label='8*x')
font = {'family':'serif','color':'blue','size':16}
plt.title("Tu nombre y apellido", loc = 'center', fontdict = font)
plt.xlabel("x")
plt.ylabel("y")
plt.vlines(2, 0, 25, linestyles ="dashed", colors ="c")
plt.vlines(-2, 0, 25, linestyles ="dashed", colors ="c")
plt.hlines(12.5, -2, 2, linestyles ="dashed", colors ="c")
plt.legend()
plt.grid()
```

Representa la función $\frac{1}{1+e^{-x}}$

<img src="sigmoide.png" ALIGN="left">

```python
# representa la función 1/(1+e^x) ,
import math

def f(x):
    return 1/(1+math.exp(-x))

x = np.arange(-10, 10, 0.1)
y2 = []
y=[]
for i in x:
    y.append(f(i))
    

plt.plot(x,y, linestyle = '-',linewidth = '1.5',color='g',label='$\\frac{1}{(1+e^{-x})}$')

#plt.plot(x,x*8,marker = '+', linestyle = '--',linewidth = '1.5',color='r',label='8*x')
font = {'family':'serif','color':'blue','size':16}
plt.title("Tu nombre y apellido", loc = 'center', fontdict = font)
plt.xlabel("x")
plt.ylabel("y")
plt.vlines(2.5, 0, 1, linestyles ="dashed", colors ="c")
plt.vlines(-2.5, 0, 1, linestyles ="dashed", colors ="c")
plt.hlines(1, -2.5, 2.5, linestyles ="dashed", colors ="c")
plt.hlines(0, -2.5, 2.5, linestyles ="dashed", colors ="c")
plt.legend()
plt.grid()
```

Haz la gráfica del analisis de frecuencias realizado del archivo "../04_RW/el_quijote.txt"  
tienes que obtener la siguiente gráfica:  
<img src="f6.png" ALIGN="left">

```python
abc="ABCDEFGHIJKLMNÑOPQRSTUVWXYZ"
f=[]
for i in abc:
    f.append(1)
N=0
archivo="../04_RW/el_quijote.txt"
text=open(archivo).readlines()
nmaxlines=len(text)
for line in text:
    for word in line.split():
        for i in word.upper():
            for j in abc:
                a=i.upper().replace("Á","A").replace("É","E").replace("Í","I").replace("Ó","O").replace("Ú","U")
                if a==j :                 
                    f[(abc.index(i))]= f[(abc.index(i))]+1 
                    N=N+1
        
for i in abc:
    f[(abc.index(i))]=f[(abc.index(i))]/N
    
#for i in abc:
#    print("%.2c  " % i,"%.3f" % f[(abc.index(i))])
    
    
import matplotlib.pyplot as plt
import numpy as np
x = []
y =[]

for i in range(len(abc)):
    x.append(abc[i])
    y.append(float(f[i]))
plt.bar(x,y)
plt.show()
```
