```python
plt.plot(x,y,marker = '*', linestyle = ':',linewidth = '2.5',color='g',label='$x^2$')
plt.plot(x,8*x,marker = '+', linestyle = '--',linewidth = '1.5',color='r',label='8*x')
plt.plot(x,50/(x+1),marker = '1', linestyle = '-.',linewidth = '1.5',color='b',label=r'$\frac{50}{x+1}$')
font = {'family':'serif','color':'darkred','size':15}
plt.title("Título", loc = 'center', fontdict = font)
plt.xlabel("x")
plt.ylabel('y = $x^2$ , 8x ,'r'$\frac{50}{x+1}$')
plt.legend()
plt.grid()
plt.show()
```

### Scatter

Para el siguiente gráfico "scatter" utizaremos un cmap, los cmap disponibles están en la siguiente página:  
https://matplotlib.org/stable/tutorials/colors/colormaps.html

```python
#gráfico de burbujas o grafico de bolas
def f(i,j):
    return (j**2)*(i**2)

x=[]
y=[]
z=[]
N=10
step=2
for i in np.arange(-N, N+step, step):
    for j in np.arange(-N, N+step, step):
        x.append(i)
        y.append(j)
        z.append(f(i,j))
        
plt.scatter(x, y, c=z, cmap='viridis',s=z, alpha=0.5)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x, y, z)
plt.show()
```

### Bars and Charts

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["Soleado", "Nublado", "Lluvia", "Nieve"])
y = np.array([70, 20, 9, 1])

plt.bar(x,y)
plt.show()
```

```python
plt.barh(x, y, height = 0.1,color = "red")
plt.show()
```

```python
plt.pie(y, labels = x)
plt.show()
```

```python
plt.pie(y, labels = x,startangle = 90)
plt.show()
```

Puedes consultar los colores en:  
https://matplotlib.org/stable/gallery/color/named_colors.html  
Su notación en Hexadecimal  
https://www.w3schools.com/colors/colors_hexadecimal.asp  

```python
ex = [0,0.3,0,0]
c = ["#E0FFFF","#AFEEEE","#48D1CC","#008080"]
plt.pie(y, labels = x,startangle = 90,explode=ex,colors=c)
plt.legend()
plt.legend(title = "Clima")
plt.show()
```
